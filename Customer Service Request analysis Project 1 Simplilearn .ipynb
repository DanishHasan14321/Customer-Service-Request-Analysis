### Above count shows that a lot of columns have same or very few deifferent records, hence we will convert them to Categorical data. Except Created Date and Closed Date. We will also check if date has incorrect timiline and correct it.

```python
exclude_columns = ['Created Date','Closed Date']

for col in exclude_columns:
    data[col] = pd.to_datetime(data[col],format='%m/%d/%Y %I:%M:%S %p')
    
for col in data.columns:
    if data[col].nunique() < 400 and col not in exclude_columns:
        data[col] = data[col].astype('category')
```

```python
data.dropna(subset=['Created Date','Closed Date','City'], inplace=True)
data['Resolution_Time'] = data['Closed Date'] - data['Created Date']
```

```python
city_comp = data.groupby(['Complaint Type','City']).size().unstack()

col_number = 5
row_number = 11
fig, axes = plt.subplots(row_number,col_number, figsize=(25,25))

for i, (label,col) in enumerate(city_comp.iteritems()):
    ax = axes[int(i/col_number), i%col_number]
    col = col.sort_values(ascending=True)[:15]
    col.plot(kind='bar', ax=ax)
    ax.set_title(label)
    
plt.tight_layout()
```

```python
borough_comp = data.groupby(['Complaint Type','Borough']).size().unstack()

col_number = 2
row_number = 3
fig, axes = plt.subplots(row_number,col_number, figsize=(25,25))

for i, (label,col) in enumerate(borough_comp.iteritems()):
    ax = axes[int(i/col_number), i%col_number]
    col = col.sort_values(ascending=True)[:15]
    col.plot(kind='bar', ax=ax)
    ax.set_title(label)
    
plt.tight_layout()
```
